```python
import os
import csv
from collections import defaultdict
```

```python
video_names = ['cpr training', 'resuscitation training']
#video_names = ['resuscitation training']
cpr_frames = defaultdict(list)
resus_frames = defaultdict(list)

print("starting to get frames")
for video_name in video_names:
    label_files = ['training_video_labels/' + f \
                   for f in os.listdir('training_video_labels') \
                   if video_name in f]
    
    #print(video_name)
    for lf in label_files:
        print(lf)
        with open(lf) as csvfile:
            reader = csv.reader(csvfile, delimiter=",")
            next(reader, None)
            for row in reader:
                video_num = int(row[0])
                print(video_num)
                
                frames = []
                for r in row[1:]:
                    ends = r.split("-")
                    if len(ends) != 2:
                        continue;
                    frames.append([*range(int(ends[0]), int(ends[1])+1)])
                
                if len(frames) == 0:
                    continue;
                
                if video_name == 'cpr training':
                    if len(cpr_frames[video_num]) == 0:
                        cpr_frames[video_num] = frames
                    else:
                        i1, i2 = 0, 0
                        cf = []
                        while i1 < len(cpr_frames[video_num]) and i2 < len(frames):
                            overlap = set(cpr_frames[video_num][i1]) & set(frames[i2])
                            if len(overlap) == 0:
                                if len(frames) > len(cpr_frames[video_num]):
                                    cf.append(frames[i2])
                                    i2 += 1
                                else:
                                    cf.append(cpr_frames[video_num][i1])
                                    i1 += 1
                            else:
                                cf.append(list(overlap))
                                i1 += 1
                                i2 += 1
                        cpr_frames[video_num] = cf
                else:
                    if len(resus_frames[video_num]) == 0:
                        resus_frames[video_num] = frames
                    else:
                        i1, i2 = 0, 0
                        cf = []
                        while i1 < len(resus_frames[video_num]) and i2 < len(frames):
                            overlap = set(resus_frames[video_num][i1]) & set(frames[i2])
                            if len(overlap) == 0:
                                if len(frames) > len(resus_frames[video_num]):
                                    cf.append(frames[i2])
                                    i2 += 1
                                else:
                                    cf.append(resus_frames[video_num][i1])
                                    i1 += 1
                            else:
                                cf.append(list(overlap))
                                i1 += 1
                                i2 += 1
                        resus_frames[video_num] = cf
                        print(cf)
                            
print("finished getting frames")
```

```python
with open("cpr_labels.txt", "w") as f:
    f.write("video segment : label # : frames ([start-end])\n")
    for vid_num in cpr_frames:
        for i in range(len(cpr_frames[vid_num])):
            output_line = str(vid_num) + " : " + str(i) + " : " + \
                            str(min(cpr_frames[vid_num][i])) + "-" + \
                            str(max(cpr_frames[vid_num][i])) + "\n"
            print(output_line)
            f.write(output_line)
```

```python
with open("resuscitation_labels.txt", "w") as f:
    f.write("video segment : label # : frames (start-end)\n")
    for vid_num in resus_frames:
        for i in range(len(resus_frames[vid_num])):
            if len(resus_frames[vid_num][i]) == 0:
                continue
            output_line = str(vid_num) + " : " + str(i) + " : " + \
                            str(min(resus_frames[vid_num][i])) + "-" + \
                            str(max(resus_frames[vid_num][i])) + "\n"
            print(output_line)
            f.write(output_line)
```

```python
import zipfile
cpr_pathname = 'training_videos/healthcare_cpr_training/'
for segment_num in cpr_frames:
    zipfile_name = 'seg' + str(segment_num) + '_'
    
    for label_num in range(len(cpr_frames[segment_num])):
        fn = zipfile_name + str(label_num) + '.zip'
        with zipfile.ZipFile(fn, 'w') as myzip:
            for frame_num in cpr_frames[segment_num][label_num]:
                frame_pathname = cpr_pathname + str(segment_num) + \
                                    '/images/image_' + str(frame_num) + '.jpg'
                fname = 'image_' + str(frame_num) + '.jpg'
                myzip.write(frame_pathname, arcname = fname)
```

```python
resus_pathname = 'training_videos/healthcare_resuscitation_training/'
for segment_num in resus_frames:
    zipfile_name = 'seg' + str(segment_num) + '_'
    
    for label_num in range(len(resus_frames[segment_num])):
        fn = zipfile_name + str(label_num) + '.zip'
        with zipfile.ZipFile(fn, 'w') as myzip:
            for frame_num in resus_frames[segment_num][label_num]:
                frame_pathname = resus_pathname + str(segment_num) + \
                                    '/images/image_' + str(frame_num) + '.jpg'
                fname = 'image_' + str(frame_num) + '.jpg'
                myzip.write(frame_pathname, arcname = fname)
```
